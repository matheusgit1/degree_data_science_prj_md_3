```python
file_path = "./matriz.txt"
```

```python
lines = 0
with open(file_path, "w", encoding="utf-8") as f:
    for elemento in m:
        if(lines < len(m)-1):
            f.write(f"{elemento}\n")
        else:
            f.write(f"{elemento}")
        lines = lines + 1
    f.close()
```
